```python
import pandas as pd
 
df=pd.read_csv('seeds.csv')
 
df.info()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
x= df.iloc[:,1:4]
print(x)
```

```python
fig =plt.figure()
ax= fig.add_subplot(projection='3d')
ax.scatter(x.Perimeter,x.Compactness,x.Kernel)
ax.set_xlabel('Perimeter')
ax.set_ylabel('Compactness')
ax.set_zlabel('Kernel')
```

```python
kmeans.fit(x)
```

```python
fig =plt.figure()
ax= fig.add_subplot(projection='3d')
ax.scatter(x.Perimeter,x.Compactness,x.Kernel,c=kmeans.labels_,cmap='rainbow')
ax.set_xlabel('Perimeter')
ax.set_ylabel('Compactness')
ax.set_zlabel('Kernel')
```
