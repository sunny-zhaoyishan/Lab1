```python
import numpy as np
array = np.random.randint(0,101,size=10)
print(array)
```

```python
mean_2 = np.mean(array)
mean_1=sum(array)/len(array)
print(mean_2)
print(mean_1)
```

```python
median_value=np.median(array)
def find_median(array):
    n = len(array)
    s = sorted(array)
    if n%2 == 0:
        return (s[n//2 - 1] + s[n//2]) / 2
    else:
        return s[n//2]
print(find_median(array))
print(median_value) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,2,3])
```

```python
x=np.linspace(0,10,50)
y1=x
y2=[i**2 for i in x ]
fig, ax= plt.subplots()
ax.set_title("y1=x,y2=x^2")
ax.grid()
ax.plot(x,y1,'--r',x,y2,'*g')
```

```python
x=np.linspace(-3,3,50)
y=x
z=np.sin(x)
fig,ax =plt.subplots()
ax = plt.axes(projection='3d')
ax.grid()
ax.plot(x,y,z)
```
